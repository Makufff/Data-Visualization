```python
import requests
import pandas as pd
from tqdm import tqdm
# x = requests.get('https://dataapi.oncb.go.th/compare/prov/2557/11', verify=False)
# print(x.json()['data'][0])
# of year
round = 0
for year in budgetYear :
  round += 1 
  print(round)
  json_array = []
  for provice in tqdm(PROV_ID) :
      x = requests.get(f'https://dataapi.oncb.go.th/compare/prov/{year}/{provice}', verify=False)
      json_array.append(x.json()['data'][0])
      df = pd.json_normalize(json_array)
  df.to_csv(root+f'Raw data/each year/{year}.csv', index=False ,encoding="utf-8")
```

## Function

```python
def showmenu():
    print('---------------- MAIN MENU :3 ----------------')
    lines = open(root + 'Main_menu.txt').read().splitlines()
    all_line = len(lines)
    for i in range(1 , all_line + 1):
        print( i , lines[i - 1] )
    menu=input(f'Please select menu option ( 1 - {all_line} ) or [ENTER] to Quit: ')
    return menu
```

```python
showmenu()
```

```python
def showgraph_bar( x , x_name ,  y , y_name , topic ):
  plt.bar(x,y)
  plt.title(topic)
  plt.ylabel(y_name)
  plt.xlabel(x_name)
  plt.show()
```

```python
showgraph_bar(np.array(df_year['ปีงบประมาณ']) , 'Fiscal year' ,df_year['จำนวนคดียาเสพติดทั้งหมด(คดี)'] , 'Total number of drug crimes(case)' , 'Total number of drug cases of each year')
```

```python
def graph_case_and_human():
  fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(10,8))

  axes[0].set_title('Number of offenders each year')
  axes[0].bar(np.array(df_year['ปีงบประมาณ']),np.array(df_year['จำนวนคดียาเสพติดทั้งหมด(คดี)']))
  axes[0].set_xlabel('Fiscal year')
  axes[0].set_ylabel('Total number of drug crimes(case)')

  axes[1].set_title("Total number of drug cases of each year")
  axes[1].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['จำนวนผู้ต้องหาทั้งหมด(คน)']),'k-D')
  axes[1].set_xlabel('Fiscal year')
  axes[1].set_ylabel('Total number of suspects (people)')

  plt.tight_layout()
  
```

```python
graph_case_and_human()
```

```python
def graph_drug():
  fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(10,8))
  axes[0].set_title('Amount of drugs each year')
  axes[0].set_xlabel('Fiscal year')
  axes[0].set_ylabel('Amount of drug (kg)')
  axes[0].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['ปริมาณกัญชา(กิโลกรัม)']),'k-D' , label = 'Amount of Cannabis (kg)' , color='red')
  axes[0].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['ปริมาณไอซ์(กิโลกรัม)']),'k-D' , label = 'Amount of Ice drug (kg)' , color='green')
  axes[0].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['ปริมาณเฮโรอีน(กิโลกรัม)']),'k-D' , label = 'Amount of Heroin (kg)' , color='blue')
  axes[0].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['ปริมาณโคเคน(กิโลกรัม)']),'k-D' , label = 'Amount of Cocaine (kg)' , color='yellow')
  axes[0].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['ปริมาณคีตามีน(กิโลกรัม)']),'k-D' , label = 'Amount of Ketamine (kg)' , color='purple')
  axes[0].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['ปริมาณยาอี(กิโลกรัม)']),'k-D' , label = 'Amount of Ecstasy (kg)' , color='pink')
  axes[0].legend()


  axes[1].set_xlabel('Fiscal year')
  axes[1].set_ylabel('Amount of drug (pills)')
  axes[1].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['ปริมาณยาบ้า(เม็ด)']),'k-D' , label = 'Amount of Amphetamine (pills)' , color='black')
  axes[1].legend()
    
  plt.tight_layout()
```

```python
graph_drug()
```

```python
def graph_therapy():
  fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(10,8))

  axes[0].set_title('The number of people receiving rehabilitation treatment each year')
  axes[0].bar(np.array(df_year['ปีงบประมาณ']),np.array(df_year['จำนวนผู้เข้าบำบัดฟื้นฟูผู้เสพ/ผู้ติดทั้งหมด(ราย)']))
  axes[0].set_xlabel('Fiscal year')
  axes[0].set_ylabel('Total number of people admitted to rehabilitation treatment for drug addicts/addicts (persons)')


  axes[1].set_title("Total number of patients")
  axes[1].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['จำนวนผู้เข้าบำบัดระบบสมัครใจ(ราย)']),'k-D', label = 'The number of patients receiving voluntary treatment (persons)' , color = 'red')
  axes[1].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['จำนวนผู้เข้าบำบัดระบบบังคับ(ราย)']),'k-D', label = 'Number of patients receiving compulsory treatment (persons)' , color = 'blue')
  axes[1].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['จำนวนผู้เข้าบำบัดระบบต้องโทษ(ราย)']),'k-D', label = 'Number of people receiving systemic therapy to be punished (persons)' , color = 'purple')
  axes[1].legend()
  axes[1].set_xlabel('Fiscal year')
  axes[1].set_ylabel('number of patients (person)')

  plt.tight_layout()
```

```python
graph_therapy()
```

```python
def graph_student():
  fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(10,8))

  axes[0].set_title('Number of early childhood children receiving teaching materials to develop brain skills')
  axes[0].bar(np.array(df_year['ปีงบประมาณ']),np.array(df_year['จำนวนผู้เข้าบำบัดฟื้นฟูผู้เสพ/ผู้ติดทั้งหมด(ราย)']))
  axes[0].set_xlabel('Fiscal year')
  axes[0].set_ylabel('Number of early childhood children receiving teaching materials to develop brain skills (EF)(person)')

  axes[1].set_title("Total Number of students in Grade 1 - Grade 6 who are immunized")
  axes[1].plot(np.array(df_year['ปีงบประมาณ']),np.array(df_year['จำนวนนักเรียน ป.1 - ป.6 ที่สร้างภูมิคุ้มกันฯ(คน)']),'k-D', label = 'Number of students in Grade 1 - Grade 6 who are immunized (person)' , color = 'red')
  axes[1].legend()
  axes[1].set_xlabel('Fiscal year')
  axes[1].set_ylabel('Number of students in Grade 1 - Grade 6 who are immunized (person)')

  plt.tight_layout()
```

```python
graph_student()
```
