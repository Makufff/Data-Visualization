## Function Show Graph

```python
# กลุ่มเพศตัวอย่าง :3

def male_and_female():
  idx = np.array(dataframe['organization'])
  female = np.array(dataframe['sex_female'])    
  male = np.array(dataframe['sex_male'])

  n = len(idx)
  r = np.arange(n)
  width = 0.5

  fig, ax = plt.subplots(figsize=(15, 6))

  ax.bar(r, female, color='pink', width=width, label='sex_male')
  ax.bar(r + width, male, color='skyblue', width=width, label='sex_female')

  ax.set_title("เพศของกลุ่มตัวอย่าง")
  ax.set_xlabel("จำนวน (คน)")

  ax.set_xticks(r + width/2)
  ax.set_xticklabels(idx, rotation=90)

  ax.legend()

  for i, v in enumerate(female):
    ax.text(i, v+0.1, str(v), color='black', fontweight='bold', ha='center')
  for i, v in enumerate(male):
    ax.text(i+width, v+0.1, str(v), color='black', fontweight='bold', ha='center')

  plt.show()
```

```python
male_and_female()
```

```python
# จำนวนตัวอย่างเพศหญิง

def female_hori():
    idx = np.array(dataframe['organization'])
    female = np.array(dataframe['sex_female'])

    fig, ax = plt.subplots(figsize=(12, 8))

    ax.barh(idx, female)

    for i, v in enumerate(female):
        ax.text(v + 0.5, i, str(v), color='black', fontweight='bold')

    ax.set_xticks(range(0, 41, 10))

    plt.title('จำนวนตัวอย่างเพศหญิง')
    plt.xlabel('จำนวน (คน)')
    plt.show()
```

```python
female_hori()
```

```python
# จำนวนตัวอย่างเพศชาย
def male_hori():
    idx = np.array(dataframe['organization'])
    male = np.array(dataframe['sex_male'])

    fig, ax = plt.subplots(figsize=(12, 8))

    ax.barh(idx, male)

    for i, v in enumerate(male):
        ax.text(v + 3, i, str(v), color='black', fontweight='bold')

    ax.set_xticks(range(0, 301, 50))

    plt.title('จำนวนตัวอย่างเพศชาย')
    plt.xlabel('จำนวน (คน)')
    plt.show()
```

```python
male_hori()
```

```python
# จำนวนตัวอย่างที่มีปริมาณแอลกอฮอล์ในเลือดไม่เกินเกณต์กฏหมายกำหนด
def limit_hori():
  idx = np.array(dataframe['organization'])
  limit = np.array(dataframe['results_limit'])

  fig, ax = plt.subplots(figsize=(12, 8))

  ax.barh(idx, limit)

  for i, v in enumerate(limit):
    ax.text(v + 0.5, i, str(v), color='black', fontweight='bold')

  ax.set_xticks(range(0, 161, 20))

  plt.title('จำนวนตัวอย่างที่มีปริมาณแอลกอฮอล์ในเลือดไม่เกินเกณฑ์กฏหมายกำหนด')
  plt.xlabel('จำนวน (คน)')
  ax.set_ylim(-0.5, len(idx) - 0.5)
  plt.show()
```

```python
limit_hori()
```

```python
# จำนวนตัวอย่างที่มีปริมาณแอลกอฮอล์ในเลือดเกินเกณต์กฏหมายกำหนด
def overlimit_hori():
    idx = np.array(dataframe['organization'])
    overlimit = np.array(dataframe['results_over_limit'])

    fig, ax = plt.subplots(figsize=(12, 8))

    ax.barh(idx, overlimit , color='orange')

    for i, v in enumerate(overlimit):
        ax.text(v + 2, i, str(v), color='black', fontweight='bold')

    ax.set_xticks(range(0, 201, 50))

    plt.title('จำนวนตัวอย่างที่มีปริมาณแอลกอฮอล์ในเลือดเกินเกณต์กฏหมายกำหนด')
    plt.xlabel('จำนวน (คน)')
    plt.show()
```

```python
overlimit_hori()
```

```python
# จำนวนตัวอย่างที่มีปริมาณแอลกอฮอล์ในเลือดตามเกณต์กฏหมายกำหนดและเกินเกณต์กฏหมายกำหนด
def limit_and_overlimit_union_hori():
  idx = np.array(dataframe['organization'])
  overlimit = np.array(dataframe['results_over_limit'])
  limit = np.array(dataframe['results_limit'])

  plt.figure(figsize=(8,10))
  plt.barh(idx, overlimit, align='center')
  plt.barh(idx, limit, color='orange', align='center', left=overlimit)
  
  # Add text annotations at the center of each bar
  for i, v in enumerate(overlimit):
      plt.text(v/2, i, str(v), color='black', fontweight='bold', ha='center', va='center')
      
  for i, v in enumerate(limit):
      plt.text(v/2 + overlimit[i], i, str(v), color='black', fontweight='bold', ha='center', va='center')
  
  plt.title('จำนวนตัวอย่างที่มีปริมาณแอลกอฮอล์ในเลือดตามเกณต์กฏหมายกำหนดและเกินเกณต์กฏหมายกำหนด')
  plt.xticks(range(0, 301, 50))
  
  plt.legend(['result_over_limit', 'result_limit'], loc='lower center' , bbox_to_anchor=(0.5, -0.1), ncol=2)
  
  plt.show()
```

```python
limit_and_overlimit_union_hori()
```

```python
# จำนวนตัวอย่างที่เข้ารับการตรวจวิเคราะห์แอลกอฮอล์ในเลือด

def testing():
    idx = np.array(dataframe['organization'])
    testing = np.array(dataframe['total testing'])

    fig, ax = plt.subplots(figsize=(12, 8))

    ax.barh(idx, testing)

    for i, v in enumerate(testing):
        ax.text(v + 0.5, i, str(v), color='black', fontweight='bold')

    ax.set_xticks(range(0, 301, 50))
    ax.set_yticks(idx)

    plt.title('จำนวนตัวอย่างที่เข้ารับการตรวจวิเคราะห์แอลกอฮอล์ในเลือด')
    plt.xlabel('จำนวน (คน)')

    plt.show()
```

```python
testing()
```

```python
# จำนวนตัวอย่างที่มีปริมาณแอลกอฮอล์ในเลือดตามเกณต์กฏหมายกำหนดและเกินเกณต์กฏหมายกำหนด
def limit_and_overlimit_multi_hori():
  idx = np.array(dataframe['organization'])
  overlimit = np.array(dataframe['results_over_limit'])
  limit = np.array(dataframe['results_limit'])

  n = len(idx)
  r = np.arange(n)
  height = 0.5

  plt.barh(r, limit, color='orange', height=height, label='result_limit')
  plt.barh(r + height, overlimit, height=height, label='result_over_limit')

  plt.title('จำนวนตัวอย่างที่มีปริมาณแอลกอฮอล์ในเลือดตามเกณต์กฏหมายกำหนดและเกินเกณต์กฏหมายกำหนด')
  plt.xlabel("จำนวน (คน)")

  plt.yticks(r + height/2, idx)
  plt.xticks(np.arange(0, 200, 20))
  
  plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.35), ncol=2)

  plt.show()
```

```python
limit_and_overlimit_multi_hori()
```
